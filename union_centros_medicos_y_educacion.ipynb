```python
import pandas as pd
```

```python
df_general=pd.read_excel('datset_general_4.xlsx')
```

```python
df_atencion_primaria=pd.read_excel('Centros_AP.xlsx')
```

```python
df_hospital=pd.read_excel('hospital.xlsx')
```

```python
# Normalizar los nombres de municipio en ambos datasets
df_general['Nombre_normalizado'] = df_general['Nombre'].str.strip().str.lower()
df_hospital['Municipio_normalizado'] = df_hospital['Municipio'].str.strip().str.lower()

# Contar cuántos hospitales hay por municipio
hospitales_por_municipio = df_hospital['Municipio_normalizado'].value_counts().reset_index()
hospitales_por_municipio.columns = ['Municipio_normalizado', 'Num_Hospitales']

# Unir el conteo al df_general
df_general = df_general.merge(
    hospitales_por_municipio,
    how='left',
    left_on='Nombre_normalizado',
    right_on='Municipio_normalizado'
)

# Rellenar NaN con 0 y convertir a enteros
df_general['Num_Hospitales'] = df_general['Num_Hospitales'].fillna(0).astype(int)

# Eliminar columnas auxiliares
df_general.drop(columns=['Nombre_normalizado', 'Municipio_normalizado'], inplace=True, errors='ignore')

# Ahora df_general tiene la columna Num_Hospitales lista
```

```python
# Normalizar los nombres para asegurar coincidencias
df_general['Nombre_normalizado'] = df_general['Nombre'].str.strip().str.lower()
df_atencion_primaria['Municipio_normalizado'] = df_atencion_primaria['MUNICIPIO'].str.strip().str.lower()

# Contar cuántos centros de atención primaria hay por municipio
ap_por_municipio = df_atencion_primaria['Municipio_normalizado'].value_counts().reset_index()
ap_por_municipio.columns = ['Municipio_normalizado', 'Num_Centros_AP']

# Unir el conteo al df_general
df_general = df_general.merge(
    ap_por_municipio,
    how='left',
    left_on='Nombre_normalizado',
    right_on='Municipio_normalizado'
)

# Rellenar NaN con 0 y convertir a enteros
df_general['Num_Centros_AP'] = df_general['Num_Centros_AP'].fillna(0).astype(int)

# eliminar columnas auxiliares
df_general.drop(columns=['Municipio_normalizado'], inplace=True, errors='ignore')
```

```python
df_general['Nombre_normalizado'] = df_general['Nombre'].str.strip().str.lower()
```

```python
# Lista ampliada de artículos y partículas comunes
articulos = [
    'el', 'la', 'los', 'las', 'lo',
    'al', 'del',
    'san', 'santa',
    'sa', 'es', 'ses', 'son',
    'a', 'de', 'en', 'do',
    'da', 'das', 'dos'
]

# Función para reordenar si empieza con artículo
def reordenar_articulo(nombre):
    if not isinstance(nombre, str):
        return ''

    nombre = nombre.strip()
    partes = nombre.split()

    if len(partes) > 1 and partes[0].lower() in articulos:
        return f"{' '.join(partes[1:])}, {partes[0]}".lower()

    return nombre.lower()
```

```python
def contar_centros_por_localidad(df, nombre_columna_resultado):
    df['Municipio_normalizado'] = df['LOCALIDAD'].apply(reordenar_articulo)
    conteo = df['Municipio_normalizado'].value_counts().reset_index()
    conteo.columns = ['Municipio_normalizado', nombre_columna_resultado]
    return conteo

```

```python
conteo_infantil = contar_centros_por_localidad(df_infantil, 'Num_Infantil')
conteo_primaria = contar_centros_por_localidad(df_primaria, 'Num_Primaria')
conteo_secundaria = contar_centros_por_localidad(df_secundaria, 'Num_Secundaria')
```

```python
df_general = df_general.merge(
    conteo_infantil,
    how='left',
    left_on='Nombre_normalizado',
    right_on='Municipio_normalizado'
)

df_general = df_general.merge(
    conteo_primaria,
    how='left',
    left_on='Nombre_normalizado',
    right_on='Municipio_normalizado'
)

df_general = df_general.merge(
    conteo_secundaria,
    how='left',
    left_on='Nombre_normalizado',
    right_on='Municipio_normalizado'
)
```

```python
df_general[['Num_Infantil', 'Num_Primaria', 'Num_Secundaria']] = df_general[[
    'Num_Infantil', 'Num_Primaria', 'Num_Secundaria'
]].fillna(0).astype(int)
```
