```python
import pandas as pd
```

```python
df_general=pd.read_excel('general_6.xlsx')
```

```python
df_general['Precio Alquiler m2 Mayo 2025'].unique()
```

```python
df_general['Precio Venta m2 Mayo 2025'].unique()
```

```python
def limpiar_precio_alquiler(valor):
    if isinstance(valor, str):
        valor = valor.replace('€/m²', '').replace('€/m2', '').replace(',', '.').strip()
    try:
        return float(valor)
    except:
        return None

df_general['Precio_Alquiler'] = df_general['Precio Alquiler m2 Mayo 2025'].apply(limpiar_precio_alquiler)
```

```python
df_general['Precio_Venta'] = pd.to_numeric(df_general['Precio Venta m2 Mayo 2025'], errors='coerce')
```

```python
# Rellenar alquiler con media provincial
df_general['Precio_Alquiler'] = df_general.groupby('Provincia')['Precio_Alquiler'].transform(
    lambda x: x.fillna(x.mean())
)

# Rellenar venta con media provincial
df_general['Precio_Venta'] = df_general.groupby('Provincia')['Precio_Venta'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
df_general.to_excel("general_7.xlsx", index=False)
```
