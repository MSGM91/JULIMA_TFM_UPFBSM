```python
import pandas as pd
```

```python
provincias=pd.read_excel('union_provincias.xlsx')
general=pd.read_excel('datset_general.xlsx')
```

```python
import pandas as pd

# Hacemos el merge especificando las columnas correctamente
general = general.merge(
    provincias[['CODIGO', 'PROVINCIA']],
    left_on='Codigo',     # En el dataframe general
    right_on='CODIGO',    # En el dataframe provincias
    how='left'
)
```

```python
print(general['Codigo'].duplicated().sum())
```

```python
duplicados = general[general['Codigo'].duplicated(keep=False)]
print(duplicados)
```
